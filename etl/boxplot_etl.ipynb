```python
import pandas as pd
import pandas as pd

df = pd.read_csv('../raw_datasets/middle_east_aggregated_data.csv', delimiter=';')

df['YEAR'] = df['WEEK'].str.split('-').str[-1].astype(int)

aggregated_df = df.groupby(['YEAR', 'COUNTRY', 'EVENT_TYPE', 'SUB_EVENT_TYPE', 'CENTROID_LATITUDE', 'CENTROID_LONGITUDE']).agg({
    'EVENTS': 'sum',
    'FATALITIES': 'sum'
}).reset_index()

aggregated_df.head()

aggregated_df.to_csv("../datasets/aggregated_map_data.csv", index=False)
```
