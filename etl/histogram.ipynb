```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
from statistics import quantiles


# Leggi il file CSV e carica in un DataFrame
df = pd.read_csv('../datasets/mea_aggregated_data.csv', delimiter=',')

# Siria + Attack
mask = (
    df['COUNTRY'].astype(str).str.contains('Syria', case=False, na=False) &
    df['SUB_EVENT_TYPE'].astype(str).str.contains('Attack', case=False, na=False)
)
syria_attack = df[mask].copy()

# Filtra dal 2014 in poi
syria_attack = syria_attack[syria_attack['YEAR'] >= 2014]

# Raggruppamento per anno con somma della colonna "EVENTS"
yearly_sum = (
    syria_attack.groupby('YEAR')['EVENTS']
    .sum()
    .reset_index(name='sum_events')
    .sort_values('YEAR')
)

# Stampa riepilogo anno per anno
print("ðŸ“Š Somma degli eventi di tipo 'Attack' in Siria (dal 2014 in poi):")
for _, row in yearly_sum.iterrows():
    print(f"  â€¢ {int(row['YEAR'])}: {row['sum_events']} eventi (totale)")

# Istogramma
plt.figure(figsize=(10,6))
plt.bar(yearly_sum['YEAR'], yearly_sum['sum_events'], color='steelblue', edgecolor='black')
plt.title("Somma degli eventi di tipo 'Attack' in Siria (dal 2014 in poi)", fontsize=14)
plt.xlabel("Anno", fontsize=12)
plt.ylabel("Somma degli eventi", fontsize=12)
plt.xticks(yearly_sum['YEAR'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
